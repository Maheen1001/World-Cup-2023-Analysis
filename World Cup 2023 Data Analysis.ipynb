## Handling missing values

```python
# Handling missing values in 'wkts': Replace with 0
data['wkts'].fillna(0, inplace=True)

# Handling missing values in 'overs', 'mdns', 'econ', '4s', '6s', 'sr', 'not_out', 'mins':
numerical_columns = ['overs', 'mdns', 'econ', '4s', '6s', 'sr', 'not_out', 'mins']

for column in numerical_columns:
    data[column].fillna(data[column].mean(), inplace=True)

print(data.isnull().sum())
```

```python
# Groupping by team and calculating team-wise performance metrics
team_performance = data.groupby('team').agg({
    'runs': 'sum',
    'wkts': 'sum',
    'bat_or_bowl': lambda x: x.mode().iloc[0],  # Most common batting/bowling style
}).reset_index()

# Identifying top-performing teams based on total runs
top_teams = team_performance.nlargest(5, 'runs')

# Print the team-wise performance metrics
print("Team-wise Performance Metrics:")
print(team_performance)

# Print the top-performing teams and their strengths
print("\nTop-performing Teams:")
print(top_teams[['team', 'runs', 'wkts', 'bat_or_bowl']])
```

```python
# Separate batting and bowling data
batting_data = data[data['bat_or_bowl'] == 'bat']
bowling_data = data[data['bat_or_bowl'] == 'bowl']

# Identify leading run-scorers
leading_run_scorers = batting_data.groupby('player')['runs'].sum().nlargest(5).reset_index()

# Identify leading wicket-takers
leading_wicket_takers = bowling_data.groupby('player')['wkts'].sum().nlargest(5).reset_index()

# Assess the impact of players on team's performance
player_impact = data.groupby('player').agg({
    'runs': 'sum',
    'wkts': 'sum',
}).reset_index()

# Print leading run-scorers and wicket-takers
print("Leading Run-scorers:")
print(leading_run_scorers)

print("\nLeading Wicket-takers:")
print(leading_wicket_takers)

# Print player impact on team's performance
print("\nPlayer Impact on Team's Performance:")
print(player_impact[['player', 'runs', 'wkts']])
```
